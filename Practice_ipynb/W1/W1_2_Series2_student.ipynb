## 16. Broadcasting
- **Broadcasting** describes the process of applying an arithmetic operation to an array (i.e., a **Series**).
- We can combine mathematical operators with a **Series** to apply the mathematical operation to every value.
- There are also methods to accomplish the same results (`add`, `sub`, `mul`, `div`, etc.)

* Broadcasting은 배열(Series) 전체에 하나의 연산을 적용하는 과정을 의미함.



```python
# <예제>

ser = pd.Series([10, 20, 30, 40])
print(ser + 5)  # 모든 요소에 5를 더함 (Broadcasting)
```

* Broadcasting 의 특징
  * Series의 모든 값에 연산이 자동 적용됨.
  * 스칼라 값(단일 값)뿐만 아니라, 배열 간 연산도 가능.
  * 연산자를 직접 사용하거나, 메서드(add(), sub(), mul(), div())등 함수 사용 가능.

```python
# <예제>
ser = pd.Series([10, 20, 30])

print(ser + 5)   # 모든 요소에 5를 더함
print(ser - 5)   # 모든 요소에서 5를 뺌
print(ser * 2)   # 모든 요소를 2배
print(ser / 2)   # 모든 요소를 2로 나눔
```

```python
# <예제>
print(ser.add(5))   # ser + 5
print(ser.sub(5))   # ser - 5
print(ser.mul(2))   # ser * 2
print(ser.div(2))   # ser / 2
```

* fill_value 옵션을 적용하여 결측값을 어떻게 처리할지 정할 수 있음.

```python
# <예제>
ser1 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
ser2 = pd.Series([5, 15], index=['a', 'c'])

print(ser1.add(ser2, fill_value=0))  # NaN을 0으로 대체 후 연산, 'b' 에 해당하는 인덱스.
```

## 17. The value_counts Method
- The `value_counts` method returns the number of times each unique value occurs in the **Series**.
- The `normalize` parameter returns the relative frequencies/percentages of the values instead of the counts.

image.png

```python
data = pd.Series(["Apple", "Banana", "Apple", "Orange", "Banana", "Apple"])
print(data.value_counts()) # 출력된 값이 고유한 값의 등장 횟수와 같은지 확인
```

* normalize 옵션을 적용해보자.

```python
print(data.value_counts(normalize=True))
```

* 기본적으로 NaN 값은 제외되지만, dropna=False로 포함 가능.


```python
data = pd.Series(["Apple", "Banana", "Apple", "Orange", "Banana", None])
print(data.value_counts(dropna=False))
```
