## Methods and Attributes between Series and DataFrames
- A **DataFrame** is a 2-dimensional table consisting of rows and columns.
- Pandas uses a `NaN` designation for cells that have a missing value. It is short for "not a number". Most operations on `NaN` values will produce `NaN` values.
- Like with a **Series**, Pandas assigns an index position/label to each **DataFrame** row.
- The **DataFrame** and **Series** have common and exclusive methods/attributes.
- The `hasnans` attribute exists only a **Series**. The `columns` attribute exists only on a **DataFrame**.
- Some methods/attributes will return different types of data.
- The `info` method returns a summary of the pandas object.

## Pandas DataFrame 개요

---

### **DataFrame의 정의**
DataFrame은 행과 열로 구성된 **2차원 테이블**임.

---

### **NaN (결측값)**
Pandas는 **결측값(missing value)**이 있는 셀에 **NaN**을 할당함.  
NaN은 *"Not a Number"*의 약어이며, 대부분의 연산에서 NaN 값을 포함하면 **결과도 NaN이 됨**.

---

### **DataFrame의 인덱스**
Series와 마찬가지로, **DataFrame의 각 행에는 인덱스 위치 또는 레이블이 할당됨**.

---

### **DataFrame과 Series의 공통점과 차이점**
DataFrame과 Series는 **일부 공통적인 메서드/속성을 가지지만, 서로 다른 속성도 존재함**.

- `hasnans` 속성은 **Series에서만 사용 가능**함.
- `columns` 속성은 **DataFrame에서만 사용 가능**함.

---

### **메서드/속성의 반환 타입 차이**
어떤 메서드나 속성은 **서로 다른 데이터 타입을 반환할 수 있음**.

## Differences between Shared Methods
- The `sum` method adds a **Series's** values.
- On a **DataFrame**, the `sum` method defaults to adding the values by traversing the index (row values).
- The `axis` parameter customizes the direction that we add across. Pass `"columns"` or `1` to add "across" the columns.

## `sum()` 메서드 개요

---

### **1. Series에서 `sum()`**
- `sum()` 메서드는 **Series의 모든 값을 더함**.

---

### **2. DataFrame에서 `sum()`**
- 기본적으로 **행(row) 방향으로 값을 더함** (즉, **각 열(column)별 합계를 계산**).

---

### **3. `axis` 매개변수 (합산 방향 조정)**
- `axis=0` 또는 `"index"` (기본값): **행을 따라 각 열의 값을 더함**.
- `axis=1` 또는 `"columns"`: **열을 따라 각 행의 값을 더함**.

## Select One Column from a DataFrame
- We can use attribute syntax (`df.column_name`) to select a column from a **DataFrame**. The syntax will not work if the column name has spaces.
- We can also use square bracket syntax (`df["column name"]`) which will work for any column name.
- Pandas extracts a column from a **DataFrame** as a **Series**.
- The **Series** is a view, so changes to the **Series** *will* affect the **DataFrame**.
- Pandas will display a warning if you mutate the **Series**. Use the `copy` method to create a duplicate.

- `df.column_name` 구문을 사용하여 `DataFrame`의 열을 선택할 수 있음. 단, 열 이름에 공백이 있으면 작동하지 않음.

- `DataFrame`에서 열을 추출하면 `Series`가 됨.
- `Series`는 뷰(view)이므로 변경 시 원본 `DataFrame`에도 영향을 미침.
- `Series`를 수정하면 Pandas가 경고를 표시할 수 있음. 복사본을 만들려면 `.copy()` 메서드를 사용.

## 🏷️ 인덱싱과 선택 (Indexing & Selection)

---

### **📌 기본적인 인덱싱 적용 방법**
DataFrame에서 데이터를 선택하는 다양한 방법을 정리함.

| 적용 방법 | 구문 | 결과 타입 |
|-----------|--------------|------------|
| **열(column) 선택** | `df[col]` | `Series` |
| **라벨(이름) 기반 행 선택** | `df.loc[label]` | `Series` |
| **정수 위치 기반 행 선택** | `df.iloc[loc]` | `Series` |
| **행 슬라이싱** | `df[5:10]` | `DataFrame` |
| **불리언 벡터를 활용한 선택** | `df[bool_vec]` | `DataFrame` |

---

### **📌 설명**
- `df[col]` → 특정 열(column)을 선택하면 `Series` 형태로 반환됨.
- `df.loc[label]` → **라벨(이름)**을 사용해 특정 행을 선택 (결과는 `Series`).
- `df.iloc[loc]` → **정수 인덱스**를 사용해 특정 행을 선택 (결과는 `Series`).
- `df[5:10]` → 행의 **슬라이싱**을 수행하면 `DataFrame`이 반환됨.
- `df[bool_vec]` → **불리언 벡터(조건 필터링)**를 사용해 특정 조건을 만족하는 행을 선택하면 `DataFrame`이 반환됨.

이 방법들을 활용하면 Pandas에서 원하는 데이터를 효율적으로 추출할 수 있음! 🚀

## Drop Rows with Missing Values
- Pandas uses a `NaN` designation for cells that have a missing value.
- The `dropna` method deletes rows with missing values. Its default behavior is to remove a row if it has *any* missing values.
- Pass the `how` parameter an argument of "all" to delete rows where all the values are `NaN`.
- The `subset` parameters customizes/limits the columns that pandas will use to drop rows with missing values.

## 결측값이 있는 행 삭제하기 (`dropna()`)

---

### **1. 결측값(NaN) 처리**
- Pandas는 **결측값(missing value)**이 있는 셀을 **NaN**으로 표시함.

---

### **2. `dropna()` 메서드**
- `dropna()`는 **결측값이 포함된 행을 삭제**하는 메서드임.
- 기본적으로 **하나라도 NaN이 포함된 행을 삭제**함.

---

### **3. `how` 매개변수**
- `how="all"`을 전달하면 **모든 값이 NaN인 행만 삭제**함.

---

### **4. `subset` 매개변수**
- `subset`을 사용하면 **특정 열만 고려하여 결측값이 있는 행을 삭제**할 수 있음.

## Rank Values with the rank Method
- The `rank` method assigns a numeric ranking to each **Series** value.
- Pandas will assign the same rank to equal values and create a "gap" in the dataset for the ranks.

## 🏆 `rank()` 메서드 - 순위 지정하기

---

### **📌 `rank()` 메서드란?**
- `rank()` 메서드는 **각 Series 값에 대해 숫자 순위를 할당**함.
- 값이 작을수록(낮을수록) **더 높은 순위(1등에 가까움)**를 가짐.

---

### **📌 동일한 값 처리**
- 같은 값이 있을 경우, **동일한 순위를 부여**하고, 그에 따라 순위 값에 **공백(gap)**이 생김.

---

## **🛠️ 예제 코드**
```python
import pandas as pd

# 샘플 데이터 생성
data = {"Name": ["Alice", "Bob", "Charlie", "David"], "Score": [85, 90, 90, 95]}
df = pd.DataFrame(data)

print("📌 원본 DataFrame:")
print(df)

# (1) 기본적인 rank() 적용
df["Rank"] = df["Score"].rank()
print("\n📌 기본 순위(Rank) 적용:")
print(df)

# (2) 내림차순으로 순위 적용
df["Rank_desc"] = df["Score"].rank(ascending=False)
print("\n📌 내림차순 순위 적용:")
print(df)
