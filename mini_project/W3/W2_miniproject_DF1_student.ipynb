## 2. .loc

* loc은 행과 열을 필터링하는 데 사용됨
* loc은 레이블 기반(label-based) 방식이므로, 선택하려는 행(또는 열)의 이름을 제공해야 함
* 참고: iloc은 "정수 인덱스 기반(integer index-based)" 방식으로, 행/열의 정수 인덱스를 사용하여 필터링할 수 있음
* 참고: "Aly"라는 성(last name)은 데이터프레임에서 한 번만 등장했기 때문에 Series가 반환됨


```python
# 다음을 실행
employee_df.loc['Aly'] # 라벨 베이스라는 의미는 무엇인가.
```


* 참고: 콜론 ":"을 사용하여 여러 행을 선택할 수 있음
* 참고: 이 방식은 **포함(inclusive)** 되므로, "Aly"와 "Patton"이 출력된 데이터프레임에 포함됨
* 주의: 정수 기반 인덱스(iloc())를 사용할 경우 결과가 다를 수 있음


```python
# 다음을 실행
employee_df.loc['Aly':'Patton'] # 라벨을 이용한 슬라이싱
```
