## Q. View 는 원본의 데이터를 참조하는 상태, Copy 는 완전히 독립된 개체를 만든것. 예제를 확인해보자.

## (1) view 예제:

```python
import pandas as pd

df = pd.DataFrame({"A": [1, 2, 3, 4, 5]})

view_series = df["A"]  # DataFrame의 한 열을 선택 → 뷰(View) 생성

print(view_series)
```

* 이 때 view_series 값을 변경하면 원본값도 변경될까? (판다스 버젼에 따라 다름). 불확실한 상황.

```python
view_series.iloc[0] = 999  # 첫 번째 값 변경
print(df)  # 원본 DataFrame 확인
```

## (2) 복사본 생성 (copy) 예제:

```python
copy_series = df["A"].copy()  # 복사본 생성
copy_series.iloc[0] = 999  # 복사본 변경

print(df)  # 원본 DataFrame 확인 (변경 없음)
```

-> 복사본은 독립적이기 때문에 변경해도 원본에 영향이 없음.
